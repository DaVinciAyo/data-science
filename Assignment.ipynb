```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
b_cancer = pd.read_csv('wisconsinBreastCancer.csv')
```

```python
b_cancer.drop(columns=['id','Unnamed: 32'], axis=1, inplace=True)
b_cancer.head()
```

```python
cancer_mean =b_cancer.iloc[:,:11]
cancer_mean.head()
```

```python
mean_heatmap = cancer_mean.drop(columns=['diagnosis'],axis=True)
```

```python
sns.heatmap(mean_heatmap.corr(), annot=True, cmap='Blues')
plt.figure(figsize=(10,10))
plt.show()
```

```python
sns.countplot(b_cancer['diagnosis'])
plt.show()
```

```python
diagnosis_count = b_cancer['diagnosis'].value_counts()
diagnosis_count
```

```python
#using the heatmap to check for the correlation between two variable and using scatterplot to check their relationship with diagnosis

plt.figure(figsize=(10,10))
sns.scatterplot(x='compactness_mean',y='concavity_mean',data=b_cancer,hue='diagnosis')
plt.title('A plot between compactness_mean vs concavity_mean showing their relationship with diagnosis')
plt.show()
```
