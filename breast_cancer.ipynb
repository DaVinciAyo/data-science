```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
b_cancer=pd.read_csv('wisconsinBreastCancer (3).csv')
```

```python
b_cancer.head()
```

```python
b_cancer.info()
```

```python
b_cancer.describe()
```

```python
b_cancer.isnull().sum()
```

```python
b_cancer.drop(columns=['Unnamed: 32','id'], inplace=True)
b_cancer.head()
```

```python
cancer_mean=b_cancer.iloc[:,:11]
cancer_mean.head()
```

```python

sns.heatmap(cancer_mean.drop(columns=['diagnosis'], axis=1).corr(),annot=True, cmap ='Blues')
plt.figure(figsize=(10,10))
plt.show
```

```python
cancer_se=b_cancer.iloc[:,11:21]
cancer_se.head()
```

```python
sns.heatmap(cancer_se.corr(),annot=True, cmap ='Reds')
plt.figure(figsize=(10,10))
plt.show
```

```python
cancer_worst=b_cancer.iloc[:,22:31]
cancer_worst.head()
```

```python
sns.heatmap(cancer_worst.corr(),annot=True, cmap ='YlOrBr')
plt.figure(figsize=(10,10))
plt.show
```

```python
sns.scatterplot(x='radius_mean',y='texture_mean',hue='diagnosis',data=b_cancer)
plt.figure(figsize=(10,10))
plt.show
```
